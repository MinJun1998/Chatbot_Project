```python
import requests
from urllib.parse import quote
import pandas as pd
```

```python
# API 키 로드
with open('keys/서울전월세key.txt') as file:
    서울_key = file.read()
```

```python
# API키를 활용해 데이터 추출
base_url = f'http://openapi.seoul.go.kr:8088/{quote(서울_key)}/json/tbLnOpendataRentV/1/500/'
results = requests.get(base_url).json()
results
```

```python
# tbLnOpendataRentV 안에있는 모든 키 확인
results['tbLnOpendataRentV'].keys()
```

```python
# 데이터 프레임으로 변환
df = pd.DataFrame(results['tbLnOpendataRentV']['row'])
df.head()
```

```python
# 필요없는 컬럼 삭제
df.drop(columns=['ACC_YEAR','SGG_CD','LAND_GBN','BJDONG_CD','LAND_GBN_NM','BOBN','BUBN','NEW_RON_SECD','CNTRCT_UPDT_RQEST_AT','BEFORE_GRNTY_AMOUNT','BEFORE_MT_RENT_CHRGE','CNTRCT_PRD'], inplace=True)
df.dropna(inplace=True)
df.tail()
```

#### 결측치 처리

```python
# 공백을 NaN으로 대체
import numpy as np
df = df.replace('', np.nan)
df
```

```python
# 결측치 확인
df.isna().sum()
```

```python
# 결측치 제거
df = df.dropna()
df
```

```python
# int 타입으로 정수 변환
df['RENT_GTN'] = df['RENT_GTN'].astype(int)
```

```python
# 답변 생성 앞부분
data = []
for i in df.index:
    result = df.SGG_NM[i]+ ' ' + df.BJDONG_NM[i] + '에 있는 ' + df.BUILD_YEAR[i] + '년에 지어진 ' + df.BLDG_NM[i] + ' '+ df.HOUSE_GBN_NM[i] + '은/는 ' + str(int(df['FLR_NO'][i])) + '층 ' + str(df['RENT_AREA'][i])
    data.append(result)
data
```

```python
# 답변 생성 뒷부분, 전세일경우 임대료 0원임으로 출력 생략
data_2 = []
for i in df.index:
    if df.RENT_FEE[i] == '0':
        result1 = 'm²(임대면적) 기준으로 보증금 ' + df['RENT_GTN'][i] + '만원 ' + df['RENT_GBN'][i] + '로 거래되었습니다.'
        data_2.append(result1)
    else:
        result2 = 'm²(임대면적) 기준으로 보증금 ' + df['RENT_GTN'][i] + '만원 임대료는 ' + df['RENT_FEE'][i] + '만원에 ' + df['RENT_GBN'][i] + '로 거래되었습니다.'
        data_2.append(result2)
data_2
```

```python
# 답변 두개 합치기
data = []
data_2 = []
for i in df.index:
    result = df.SGG_NM[i]+ ' ' + df.BJDONG_NM[i] + '에 있는 ' + df.BUILD_YEAR[i] + '년에 지어진 ' + df.BLDG_NM[i] + ' '+ df.HOUSE_GBN_NM[i] + '은/는 ' + str(int(df['FLR_NO'][i])) + '층 ' + str(df['RENT_AREA'][i])
    data.append(result)
    for i in df.index:
        if df.RENT_FEE[i] == '0':
            result1 = 'm²(임대면적) 기준으로 보증금 ' + df['RENT_GTN'][i] + '만원 ' + df['RENT_GBN'][i] + '로 거래되었습니다.'
            data_2.append(result1)
        else:
            result2 = 'm²(임대면적) 기준으로 보증금 ' + df['RENT_GTN'][i] + '만원 임대료는 ' + df['RENT_FEE'][i] + '만원에 ' + df['RENT_GBN'][i] + '로 거래되었습니다.'
            data_2.append(result2)
data_3 = [a + b for a, b in zip(data, data_2)]
data_3
```

```python
# 데이터프레임 생성 및 컬럼 3개 생성
home = pd.DataFrame()
home['구분'] = ''
home['질문'] = data_3
home['답변'] = data_3
```
