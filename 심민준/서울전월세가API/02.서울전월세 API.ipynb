```python
import requests
from urllib.parse import quote
import pandas as pd
```

```python
# 데이터 프레임으로 변환
df = pd.read_csv("data/서울전월세_전처리.csv")
df.head(3)
```

```python
# int 타입으로 정수 변환
df['RENT_GTN'] = df['RENT_GTN'].astype(int)
```

```python
# 답변 생성 앞부분
data = []
for i in df.index:
    result = f'{df.SGG_NM[i]} {df.BJDONG_NM[i]}에 있는 {df.BUILD_YEAR[i]}년에 지어진 {df.BLDG_NM[i]} {df.HOUSE_GBN_NM[i]}은/는 {df.FLR_NO[i]}층 {df.RENT_AREA[i]}'
    data.append(result)
data
```

```python
# 답변 생성 뒷부분, 전세일경우 임대료 0원임으로 출력 생략
data_2 = []
for i in df.index:
    if df.RENT_FEE[i] == 0:
        result1 = f'm²(임대면적) 기준으로 보증금 {df.RENT_GTN[i]}만원 {df.RENT_GBN[i]}로 거래되었습니다.'
        data_2.append(result1)
    else:
        result2 = f'm²(임대면적) 기준으로 보증금 {df.RENT_GTN[i]}만원 임대료는 {df.RENT_FEE[i]}만원에 {df.RENT_GBN[i]}로 거래되었습니다.'
        data_2.append(result2)
data_2
```

```python
# 질문 만들기
data_q = []
for i in df.index:
    result = f'{df.SGG_NM[i]} {df.BJDONG_NM[i]}에 {df.HOUSE_GBN_NM[i]} {df.RENT_GBN[i]} 보여주세요.'
    data_q.append(result)
data_q
```

```python
# 답변 두개 합치기
data = []
data_2 = []
for i in df.index:
    result = f'{df.SGG_NM[i]} {df.BJDONG_NM[i]}에 있는 {df.BUILD_YEAR[i]}년에 지어진 {df.BLDG_NM[i]} {df.HOUSE_GBN_NM[i]}은/는 {df.FLR_NO[i]}층 {df.RENT_AREA[i]}'
    data.append(result)
    for i in df.index:
        if df.RENT_FEE[i] == 0:
            result1 = f'm²(임대면적) 기준으로 보증금 {df.RENT_GTN[i]}만원 {df.RENT_GBN[i]}로 거래되었습니다.'
            data_2.append(result1)
        else:
            result2 = f'm²(임대면적) 기준으로 보증금 {df.RENT_GTN[i]}만원 임대료는 {df.RENT_FEE[i]}만원에 {df.RENT_GBN[i]}로 거래되었습니다.'
            data_2.append(result2)
data_3 = [a + b for a, b in zip(data, data_2)]
data_3
```

```python
# 구분에 들어갈 내용 테스트
data_2 = []
data = []
data = str(df.SGG_NM[0] + df.BJDONG_NM[0] + df.RENT_GBN[0])
data_2.append({'데이터': data})
data_2
```

```python
# 구분안에 들어갈 내용
data_2 = []
for i in df.index:
    data = []
    data = str(df.SGG_NM[i] + ' ' + df.BJDONG_NM[i] + ' ' + df.RENT_GBN[i])
    data_2.append(data)
data_2
```

```python
# 데이터프레임 생성 및 컬럼 3개 생성
home = pd.DataFrame()
home['구분'] = data_2
home['질문'] = data_q
home['답변'] = data_3
```
