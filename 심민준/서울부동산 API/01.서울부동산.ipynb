```python
import requests
from urllib.parse import quote
import pandas as pd
```

```python
# API 키 로드
with open('key/서울부동산key.txt') as file:
    부동산_key = file.read()
```

```python
a = 1
b = 1000
data = []
for i in range(26):
    base_url = f'http://openapi.seoul.go.kr:8088/{quote(부동산_key)}/json/landBizInfo/{a}/{b}/'
    url = requests.get(base_url).json()
    df = pd.DataFrame(url['landBizInfo']['row'])
    data.append(df)
    a += 1000
    b += 1000
df1 = pd.concat(data, ignore_index=True)
```

```python
df1.drop(columns=['SYS_REGNO', 'SGG_CD', 'BJDONG_CD', 'LAND_GBN', 'BOBN', 'BUBN', 'BEGIN_DE', 'END_DE', 'RNUM', 'ROAD_CD', 'BLDG_LOC', 'BLDG_BOBN', 'BLDG_BUBN', 'STS_GBN', 'SGG_NM', 'BJDONG_NM'], inplace=True)
df1.dropna(inplace=True)
df1.tail()
```

```python
# csv파일로 저장
df1.to_csv('data/서울부동산.csv', index=False, encoding='utf-8')
```
