```python
import requests
from urllib.parse import quote
import pandas as pd
```

```python
# API 키 로드
with open('key/서울부동산key.txt') as file:
    부동산_key = file.read()
```

```python
# 한번에 1000개 데이터밖에 불러오지못해서 반복문을 사용해 데이터 불러오기
a = 1
b = 1000
data = []
for i in range(26):
    base_url = f'http://openapi.seoul.go.kr:8088/{quote(부동산_key)}/json/landBizInfo/{a}/{b}/'
    url = requests.get(base_url).json()
    df = pd.DataFrame(url['landBizInfo']['row'])
    data.append(df)
    a += 1000
    b += 1000
base_url1 = f'http://openapi.seoul.go.kr:8088/{quote(부동산_key)}/json/landBizInfo/26001/26608/'
url1 = requests.get(base_url1).json()
df1 = pd.DataFrame(url1['landBizInfo']['row'])
data.append(df1)
df2 = pd.concat(data, ignore_index=True)
```

```python
# 필요없는 컬럼 제거
df2.drop(columns=['SYS_REGNO', 'SGG_CD', 'BJDONG_CD', 'LAND_GBN', 'BOBN', 'BUBN', 'BEGIN_DE', 'END_DE', 'RNUM', 'ROAD_CD', 'BLDG_LOC', 'BLDG_BOBN', 'BLDG_BUBN', 'STS_GBN', 'SGG_NM', 'BJDONG_NM'], inplace=True)
df2.dropna(inplace=True)
df2.tail()
```
