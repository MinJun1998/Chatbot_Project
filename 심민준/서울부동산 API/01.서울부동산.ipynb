```python
import requests
from urllib.parse import quote
import pandas as pd
```

```python
# API 키 로드
with open('key/서울부동산key.txt') as file:
    부동산_key = file.read()
```

```python
# API키를 활용해 데이터 추출
base_url = f'http://openapi.seoul.go.kr:8088/{quote(부동산_key)}/json/landBizInfo/1/1000/'
url1 = requests.get(base_url).json()
```

```python
a = 1
b = 1000
data = []
for i in range(2):
    base_url = f'http://openapi.seoul.go.kr:8088/{quote(부동산_key)}/json/landBizInfo/{a}/{b}/'
    url = requests.get(base_url).json()
    df = pd.DataFrame(url['landBizInfo']['row'])
    data = data.append(df)
```

```python
# 데이터 프레임으로 변환
df = pd.DataFrame(url1['landBizInfo']['row'])
df.head()
```

```python
df.keys()
```

```python
df.drop(columns=['SYS_REGNO', 'SGG_CD', 'BJDONG_CD', 'LAND_GBN', 'BOBN', 'BUBN', 'BEGIN_DE', 'END_DE', 'RNUM', 'ROAD_CD', 'BLDG_LOC', 'BLDG_BOBN', 'BLDG_BUBN', 'STS_GBN', 'SGG_NM', 'BJDONG_NM'], inplace=True)
df.dropna(inplace=True)
df.tail()
```
