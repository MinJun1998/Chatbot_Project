```python
import requests
from urllib.parse import quote
import pandas as pd
```

```python
# API 키 로드
with open('keys/서울전월세key.txt') as file:
    서울_key = file.read()
```

```python
# API키를 활용해 데이터 추출
base_url = f'http://openapi.seoul.go.kr:8088/{quote(서울_key)}/json/tbLnOpendataRentV/1/500/'
results = requests.get(base_url).json()
results
```

```python
# tbLnOpendataRentV 안에있는 모든 키 확인
results['tbLnOpendataRentV'].keys()
```

```python
# 데이터 프레임으로 변환
df = pd.DataFrame(results['tbLnOpendataRentV']['row'])
df.head()
```

```python
# 필요없는 컬럼 삭제
df.drop(columns=['ACC_YEAR','SGG_CD','LAND_GBN','BJDONG_CD','LAND_GBN_NM','BOBN','BUBN','NEW_RON_SECD','CNTRCT_UPDT_RQEST_AT','BEFORE_GRNTY_AMOUNT','BEFORE_MT_RENT_CHRGE','CNTRCT_PRD'], inplace=True)
df.dropna(inplace=True)
df.tail()
```

```python
data = []
for i in df.index:
    result = df.SGG_NM[i]+ ' ' + df.BJDONG_NM[i] + '에 있는 ' + df.BUILD_YEAR[i] + '년에 지어진 ' + df.BLDG_NM[i] + ' '+ df.HOUSE_GBN_NM[i] + '은/는 ' + str(int(df['FLR_NO'][i])) + '층 ' + str(df['RENT_AREA'][i])
    data.append(result)
data
```

```python
data_1 = []
data_1 = 'm²(임대면적) 기준으로 보증금 ' + df['RENT_GTN'] + '만원 임대료는 ' + df['RENT_FEE'] + '만원에 ' + df['RENT_GBN'] + '로 거래되었습니다.' 
data_1
```

```python
answer = data + data_1
answer
```

```python
home = pd.DataFrame()
home['구분'] = '전월세가'
home['질문'] = answer
home['답변'] = answer
```
