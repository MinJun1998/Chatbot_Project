```python
import requests
from urllib.parse import quote
import pandas as pd
```

```python
# API 키 로드
with open('keys/서울전월세key.txt') as file:
    서울_key = file.read()
```

```python
# API키를 활용해 데이터 추출
base_url = f'http://openapi.seoul.go.kr:8088/{quote(서울_key)}/json/tbLnOpendataRentV/1/10000/'
results = requests.get(base_url).json()
results
```

```python
# tbLnOpendataRentV 안에있는 모든 키 확인
results['tbLnOpendataRentV'].keys()
```

```python
# 데이터 프레임으로 변환
df = pd.DataFrame(results['tbLnOpendataRentV']['row'])
df.head()
```

```python
# 필요없는 컬럼 삭제
df.drop(columns=['ACC_YEAR','SGG_CD','LAND_GBN','BJDONG_CD','LAND_GBN_NM','BOBN','BUBN','NEW_RON_SECD','CNTRCT_UPDT_RQEST_AT','BEFORE_GRNTY_AMOUNT','BEFORE_MT_RENT_CHRGE','CNTRCT_PRD'], inplace=True)
df.dropna(inplace=True)
df.tail()
```

#### 결측치 처리

```python
# 공백을 NaN으로 대체
import numpy as np
df = df.replace('', np.nan)
df
```

```python
# 변경 확인
df.iloc[496]
```

```python
# 결측치 확인
df.isna().sum()
```

```python
# 결측치 제거
df = df.dropna()
df
```

```python
# csv파일로 저장
df.to_csv('data/서울전월세_전처리.csv', index=False, encoding='utf-8')
```
