```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/서울전월세_전처리.csv')
df
```

```python
# SGG_NM컬럼을 셀렉해서 중복을 제외한 값
all_dong = df[['BJDONG_NM']]
print(all_dong.shape, all_dong.BJDONG_NM.nunique())
```

```python
# SGG_NM컬럼 중복제거
all_dong.drop_duplicates(subset=['BJDONG_NM'], inplace=True)
all_dong.shape
```

```python
# 뒤죽박죽 되어있는 인덱스 리셋
all_dong = all_dong.reset_index()
list_dong = all_dong['BJDONG_NM']
```

```python
# 리스트동의 인덱스
list_dong.index
```

```python
# 출력테스트
list_dong[2]
```

```python
#모든동의 보증금이 가장비싼 전세 반복
last = []
question = []
answer = []
end = []
for i in list_dong.index:
    max = []
    input = df[df.BJDONG_NM == list_dong[i]]
    input = input[input.RENT_GBN == '전세'] # 전세or월세
    try:
        max = input['RENT_GTN'].idxmax()    # RENT
        max = input.loc[max]
        last.append(max)
    except:
        answer.append(list_dong[i] + '의 (전세)거래내역이 없습니다')
        question.append(list_dong[i] + ' 전세' + ' 보증금' + ' 가장비싼' + ' 높은' + ' 고액')
dcm = pd.DataFrame(last)
dcm = dcm.reset_index()
dcm.drop(columns=['index'], inplace=True)
dcm

for i in dcm.index:
    question.append(dcm.BJDONG_NM[i] + ' 전세' + ' 보증금' + ' 가장비싼' + ' 높은' + ' 고액')

for i in dcm.index:
    answer.append(dcm.BJDONG_NM[i] + '에서 보증금이 가장비싸게 거래된 ' + dcm.RENT_GBN[i] + '는 m²(임대면적) 기준으로 보증금 ' + str(dcm['RENT_GTN'][i]) + '만원에 거래된 ' + dcm.BLDG_NM[i] + ' ' + dcm.HOUSE_GBN_NM[i] + '입니다.')
    
for i in dcm.index:
    end.append({'구분':None,'질문':question[i], '답변':answer[i]})
```

```python
#모든동의 임대료가 가장싼 월세 반복
last = []
answer = []
question = []
for i in list_dong.index:
    max = []
    input = df[df.BJDONG_NM == list_dong[i]]
    input = input[input.RENT_GBN == '월세'] # 전세or월세
    try:
        max = input['RENT_FEE'].idxmax()    # RENT_GTN = 보증금, RENT_FEE = 임대료
        max = input.loc[max]
        last.append(max)
    except:
        answer.append(list_dong[i] + '의 (월세)거래내역이 없습니다')
        question.append(list_dong[i] + ' 월세' + ' 임대료' + ' 가장비싼' + ' 높은' + ' 고액')
dcm = pd.DataFrame(last)
dcm = dcm.reset_index()
dcm.drop(columns=['index'], inplace=True)
dcm
for i in dcm.index:
    question.append(dcm.BJDONG_NM[i] + ' 월세' + ' 임대료' + ' 가장비싼' + ' 높은' + ' 고액')
    
for i in dcm.index:
    answer.append(dcm.BJDONG_NM[i] + '에서 임대료가 가장비싸게 거래된 ' + dcm.RENT_GBN[i] + '는 m²(임대면적) 기준으로 보증금 ' + str(dcm['RENT_GTN'][i]) + '만원 임대료는 ' + str(dcm['RENT_FEE'][i]) + '만원에 거래된 ' + dcm.BLDG_NM[i] + ' ' + dcm.HOUSE_GBN_NM[i] + '입니다.')

for i in dcm.index:
    end.append({'구분':None,'질문':question[i], '답변':answer[i]})
```
