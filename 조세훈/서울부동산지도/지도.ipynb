```python
import pandas as pd
df = pd.read_csv('data/서울부동산.csv')
df.head()
```

```python
df.BJDONG_NM.str.find('구로동') 
```

```python
gf = df[df.BJDONG_NM.str.find('구로동') == 0.0]
```

```python
import folium
map = folium.Map([gf.위도.mean(), gf.경도.mean()], zoom_start=13)
map
```

```python
map = folium.Map([gf.위도.mean(), gf.경도.mean()], zoom_start=14)
for i in gf.index:
    folium.Marker([gf.위도[i], gf.경도[i]], tooltip= [gf.CMP_NM[i], '<br>', gf.ADDRESS[i],'<br>', gf.RA_REGNO[i]]).add_to(map)

map
```

```python
def map_by_dong(df, dong):
    gf = df[df.BJDONG_NM.str.find(dong) == 0.0] 
    map = folium.Map([gf.위도.mean(), gf.경도.mean()], zoom_start=15)
    for i in gf.index:
        folium.Marker([gf.위도[i], gf.경도[i]], tooltip=[gf.CMP_NM[i],'\n', gf.ADDRESS[i], gf.RA_REGNO[i]]).add_to(map)
    map.save('dong.html')
```

```python
map_by_dong(df, '신도림동')
```
