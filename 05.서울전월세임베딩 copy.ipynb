```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
# 
model = SentenceTransformer('jhgan/ko-sroberta-multitask')
```

```python
df = pd.read_csv('data/서울전월세.csv')
df.head()
```

```python
df.isna()
```

```python
df = df.dropna()
df
```

```python
df.shape
```

- 인코딩 후 데이터 셋 저장

```python
model.encode(df.질문[0])
```

```python
df['embedding'] = pd.Series([[]] * len(df))     #  더미데이터
df.shape
```

```python
df.embedding = df.질문.map(lambda x: list(model.encode(x)))
df.head(3)
```

```python
df.to_csv('data/서울전월세_dataset.csv', index=False)
```

- 챗봇

```python
import json
wdf = pd.read_csv('data/서울전월세_dataset.csv')
wdf.embedding = wdf.embedding.apply(json.loads)
wdf.head(3)
```

```python
text = '서울 월세 싼곳 추천해주세요'
embedding = model.encode(text)
```

```python
wdf['유사도'] = wdf.embedding.map(lambda x: cosine_similarity([embedding], [x]).squeeze())
wdf.head(3)
```

```python
wdf.유사도.idxmax()
```

```python
wdf.iloc[wdf.유사도.nlargest(5)]
```

```python
answer = 5
sr = [wdf['답변'][i] for i in wdf['유사도'].argsort()[:-answer-1:-1]]
```

```python
for i, response in enumerate(sr, 1):
    print(f"{i}: {response}")
```

```python
wdf.답변[wdf.유사도.idxmax()]
```

```python
text = '영등포에서 임대료가 가장 싼 곳'
embedding = model.encode(text)
wdf['유사도'] = wdf.embedding.map(lambda x: cosine_similarity([embedding], [x]).squeeze())
answer = 5
sr = [wdf['답변'][i] for i in wdf['유사도'].argsort()[:-answer-1:-1]]
for i, response in enumerate(sr, 1): print(f"{i}: {response}")
```
